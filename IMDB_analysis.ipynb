```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/IMDB-Movie-Data.csv")
df.head()
```

```python
df.shape  #getting structure
```

```python
df.info()  #getting information
```

```python
df.isnull().sum() #checking missing values
```

```python
precentage_missing = df.isnull().sum()*100/len(df)
print(precentage_missing)   #percentage of missing value
```

```python
df.dropna(axis=0)
```

```python
#checking for duplicate data
dup_data=df.duplicated().any()
print(dup_data)
```

```python
#to get satistics (mean, median, std, min, max)
df.describe()
```

```python
# movie having runtime >=180min
df[df['Runtime (Minutes)']>=180]['Title']
```

```python
# in which year have highest avg voting
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year', y='Votes', data=df)
plt.title("Average votes")
plt.show()
```

```python
# year having highest, average  revenue
df.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year', y='Revenue (Millions)', data=df)
plt.title("Average revenue")
plt.show()
```

```python
# avg rating for each director
df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

```python
# top 10 lengthly movies
top=df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']].set_index('Title')
print(top)
```

```python
sns.barplot(x='Runtime (Minutes)',y=top.index,data=top)
plt.show()
```

```python
# to get no.of movies in a year
df['Year'].value_counts()
```

```python
sns.countplot(x='Year', data=df)
plt.show()
```

```python
# to find most popular movie
df[df['Revenue (Millions)'].max()==df['Revenue (Millions)']]
```
